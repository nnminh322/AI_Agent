```python
from smolagents import LiteLLMModel
```

```python
model = LiteLLMModel(
    model_id="ollama_chat/qwen2.5:1.5b",
    api_base="http://127.0.0.1:11434",
    cttx=8192,
)
```

```python
response = model.generate(messages=[{"role": "user", "content": [{"type": "text", "text": "Bạn là ai?"}]}])

# The response is a string, so you can just assign it directly
model_response = response

print(f"Phản hồi của mô hình: {model_response}")
```
